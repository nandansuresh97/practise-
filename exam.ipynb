```python

numeric_columns = df.select_dtypes(include=['number']).columns  # Identify numeric columns

# Fill missing values with median for numerical columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())
```

```python
print(df.isnull().sum())
```

```python

non_numeric_columns = df.select_dtypes(include=['object']).columns # Identify non numeric columns
missing_values_non_numeric = df[non_numeric_columns].isnull().sum()
print("Missing values in non-numeric columns:")
print(missing_values_non_numeric)
```

```python
# Fill missing values in non numeric with the mode
for col in non_numeric_columns:
    mode_value = df[col].mode()[0]
    df[col].fillna(mode_value, inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df = pd.get_dummies(df, drop_first=True) # Encoding non numeric with one-hot encoding
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

# Define features and target
X = df.drop('selection', axis=1)
y = df['selection']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the features
X_scaled = scaler.fit_transform(X)
```

```python
#RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=10, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error, r2_score
print("Mean Squared Error of the model is ",mean_squared_error(y_pred,y_test))
print("Mean absolute Error of the model is ",mean_absolute_error(y_pred,y_test))
print("R2 score of the model is ",r2_score(y_pred,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

# Best parameters and score
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```
